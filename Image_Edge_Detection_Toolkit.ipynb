```python
# Importing necessary libraries

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read an image in grayscale mode

img=cv2.imread("/content/pandas.jpeg", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap="gray")
plt.show()
```

```python
# Apply the Sobel operator to compute the horizontal derivative

sobelx= cv2.Sobel(img,-1,1,0,ksize=3)
```

```python
# Display the image

plt.imshow(sobelx, cmap="gray")
plt.show()
```

```python
# Compute the vertical derivative


sobely = cv2.Sobel(img, -1,0,1, ksize= 3)
plt.imshow(sobely, cmap="gray")
plt.show()
```

```python
# Apply the Laplacian operator for edge detection.


lap=cv2.Laplacian(img, -1)
plt.imshow(lap, cmap="gray")
plt.show()
```

```python
# Apply the Canny edge detection algorithm.


can=cv2.Canny(img, 100,200)
plt.imshow(can, cmap="gray")
plt.show()
```
