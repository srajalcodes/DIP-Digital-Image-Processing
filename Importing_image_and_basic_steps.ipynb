```python
# Importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image in grayscale mode
image = cv2.imread("/content/Fish.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(image)
plt.show()
```
