```python
# Importing necessary libraries

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read an image

image= cv2.imread("/content/Unknown.jpeg")
plt.imshow(image)
plt.show()
```

```python
# Apply a 2x2 blur filter to the image using OpenCV.


blurred_img=cv2.blur(image,(2,2))
plt.imshow(blurred_img)
plt.show()
```

```python
# Read the image in grayscale mode

image= cv2.imread("/content/Unknown.jpeg",cv2.IMREAD_GRAYSCALE)
plt.imshow(image)
plt.show()
```
