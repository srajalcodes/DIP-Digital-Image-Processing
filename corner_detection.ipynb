```
import cv2
import numpy as np
from matplotlib import pyplot as plt
img=cv2.imread("/content/images.jpeg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
plt.show()
```

```
corners = cv2.goodFeaturesToTrack(img,1, 0.01, 1)
corners = np.int0(corners)
for i in corners:
  x,y=i.ravel()
  cv2.circle(img,(x,y),1, 1)
  plt.imshow(img, cmap= "gray")
  plt.show()
```
