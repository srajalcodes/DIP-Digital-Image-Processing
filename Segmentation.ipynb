Global Thresholding

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image= cv2.imread("/content/lion.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
plt.show()
```

```python
ret,img1=cv2.threshold(image,80,255,cv2.THRESH_BINARY)
plt.imshow(img1, cmap="gray")
plt.show()
```

```python
ret,img2=cv2.threshold(image,80,255,cv2.THRESH_BINARY_INV)
plt.imshow(img2, cmap="gray")
plt.show()
```

Local Thresholding

```python
local_thresh=cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
plt.imshow(local_thresh,cmap="gray")
plt.show()
```

Otsu method

```python
ret,  img3=cv2.threshold(image,0,255,cv2.THRESH_OTSU)
plt.imshow(img3, cmap="gray")
plt.show()
```
